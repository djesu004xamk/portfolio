Let's import pandas, sklearn datasets and numpy. We will use local dataset "Cancer_Data.csv" containing data about benign and malignant  cancers

```python
from sklearn import datasets
import pandas as pd
import numpy as np
cancer = pd.read_csv('Cancer_Data.csv')
cancer
```

```python
print(cancer.diagnosis)
```

Let's check if there are empty cells

```python
cancer.isnull().sum()
```

Now we shall change our diagnosis code to numeric value ->
0 = Malignant
1 = Benign

```python
def type_to_numeric(x):
        if x=='M': return 0
        if x=='B':   return 1
```

And after that, we'll fill epmty cells with a value. We'll also put diagnosis code to a variable y. Rest of the table will be available in variable x. We shall also remove id and Unnamed: 32 -columns because they they nothing to the training. Diagnosis column were also removed it including the target values.

```python
from sklearn import preprocessing

cancer.fillna(0, inplace=True)

y = cancer['diagnosis'].apply(type_to_numeric)

cancer.drop(['id', 'diagnosis', 'Unnamed: 32'],axis='columns',inplace=True)

x = cancer

x
```

Let's divide dataset into training (80%) and testing (20%) groups 

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

Let's train with 5 neighbours

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(x_train, y_train)
```

Here we can see the accuracy of our model.

```python
knn.score(x_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
scores = cross_val_score(KNeighborsClassifier(n_neighbors=5), x, y,cv=3)
np.average(scores)
```

We can see how the model was able to predict the right target with confusion matrix

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
cm
```
